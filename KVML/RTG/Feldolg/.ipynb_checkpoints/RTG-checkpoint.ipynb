# M√°sodik feladat

```python
data=loadtxt('./AlMg.dat')
```

```python
figsize(8,6)
plt.semilogy(data[:,0],data[:,1])
grid(True)
xlabel('2$\Theta$ [¬∞]',fontsize=13)
ylabel('Be√ºt√©ssz√°m',fontsize=13)
title('Al(Mg) r√∂ntgendiffrakci√≥s spektruma',fontsize=15)
#xlim(25,145)
savefig('./AlMg.png',bbox_inches='tight')
```

```python
a=[0.4063276434, \
    0.4073586933, \
    0.4067611852, \
    0.4071334724, \
    0.4069909216, \
    0.4070409884, \
    0.4074937006, \
    0.4069076186, \
    0.4073794206]

F=[5.3785601487,\
4.4708061257,\
2.5891209302,\
1.8563267754,\
1.6635560581,\
1.0581586001,\
0.7181177326,\
0.6121448931,\
0.2701131699]
```

```python
def linear(x,a,b):
    return(a*x+b)
```

```python
popt, pcov = curve_fit(linear,F,a)
perr=sqrt(diag(pcov))
```

```python
print(popt, '\n', perr)
```

```python
figsize(10,6)
scatter(F,a,color='navy',label='M√©r√©si pontok')
km=array([0,6])
plot(km,linear(km,*popt),label='Illesztett egyenes \n y(x)=$(-1.09 \cdot 10^{-4}) \cdot x + 0.4073$')
xlim(0,6)
ylim(0.405,0.409)
grid(True)
title('R√°csparam√©ter meghat√°roz√°sa',fontsize=15)
xlabel('F($\Theta$)', fontsize=13)
ylabel('a$_{hkl}$ [nm]', fontsize=13)
legend(fontsize=10)
savefig('./otvoz.png',bbox_inches='tight')
```

```python
kalib=loadtxt('AlMg_konc.dat')
```

```python
popt, pcov = curve_fit(linear,kalib[:,1],kalib[:,0])
perr=sqrt(diag(pcov))
```

```python
print(popt, '\n', perr)
```

```python
km=array([-1,2])
plot(km,linear(km,*popt),label='Illesztett egyenes \n $y(x)=2288.70 \cdot x - 926.59$')
scatter(kalib[:,1],kalib[:,0],color='navy',label='Kalibr√°ci√≥s pontok')
xlim(0.404, 0.409)
ylim(-1,9)
grid(True)
title('Kalibr√°ci√≥s egyenes',fontsize=15)
xlabel('$a_0$ [nm]',fontsize=13)
ylabel('Mg atomsz√°zal√©k',fontsize=13)
legend(fontsize=10)
savefig('./kalib.png',bbox_inches='tight')
```
